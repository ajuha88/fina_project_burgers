```python
import requests, json
import pandas as pd
from config import client_id, client_secret
```

```python
burgers_df = pd.read_csv("burger_joints.csv")
burgers_df = burgers_df.drop("Unnamed: 0", axis=1)
burgers_df.head()
```

```python
CLIENT_ID = client_id
CLIENT_SECRET = client_secret
VERSION = "20180605"
```

```python
all_likes = []

for row in burgers_df.iterrows():
    url =  f'https://api.foursquare.com/v2/venues/{row[1][3]}/likes/'
    
    params = dict(
        client_id=CLIENT_ID,
        client_secret=CLIENT_SECRET,
        v=VERSION,
    )

    response = requests.get(url=url, params=params).json()
    num_likes = response['response']['likes']['count']
#     print(f"{row[1][4]}'s Likes: {num_likes}")
    
    if num_likes != 0:
        response = response['response']['likes']['items']

        for likes in response:
            likes['Venue ID'] = row[1][3]
            likes['Venue'] = row[1][4]
            likes['Venue Category'] = row[1][5]
            likes['Venue Latitude'] = row[1][6]
            likes['Venue Longitude'] = row[1][7]

        
        all_likes.append(response)
```

```python
likes_list = []

for i in all_likes:
    for x in i:
        likes_list.append(x)

likes_df = pd.DataFrame(likes_list)
likes_df.head()
```

```python
likes_df = likes_df.drop(["photo", "type", "venue"], axis=1)
```
