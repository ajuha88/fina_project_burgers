```python
import pandas as pd
```

```python
merged_df = pd.read_csv("merged_data.csv")
merged_df = merged_df.drop("Unnamed: 0", axis=1)
merged_df.head()
```

```python
count_df = merged_df.groupby("Neighborhood", as_index=False).count()[['Neighborhood', 'Venue']]
count_df.head()
```

```python
avg_df = merged_df.groupby("Neighborhood", as_index=False).mean()[['Neighborhood', 'Neighborhood Latitude',
                                                                   'Neighborhood Longitude', 'Rating']]
avg_df.head()
```

```python
sum_df =  merged_df.groupby("Neighborhood", as_index=False).sum()[['Neighborhood', 'Tip Count', 'Like Count']]
sum_df.head()
```

```python
neighborhoods_df = count_df.merge(avg_df, how="inner", on="Neighborhood").merge(sum_df, how="inner", on="Neighborhood")
neighborhoods_df.head()
```

```python
neighborhoods_df.to_csv("by_neighborhood.csv")
```
