```python
import pandas as pd
```

```python
venues_df = pd.read_csv("burger_joints.csv")
likes_df = pd.read_csv("likes.csv")
ratings_df = pd.read_csv("ratings.csv")
tips_df = pd.read_csv("tips.csv")
```

```python
venues_df = venues_df.drop("Unnamed: 0", axis=1)
likes_df = likes_df.drop("Unnamed: 0", axis=1)
ratings_df = ratings_df.drop("Unnamed: 0", axis=1)
tips_df = tips_df.drop("Unnamed: 0", axis=1)
```

```python
tips_df.head()
```

```python
merge1_df = venues_df.merge(tips_df, how='inner', on=["Venue ID"])
merge1_df.head()
```

```python
merge2_df = merge1_df.merge(ratings_df, how='inner', on=["Venue ID"])
merge2_df.head()
```

```python
merge2_df.to_csv("merged_data.csv")
```

```python
grouped_likes = likes_df.groupby("Venue ID", as_index=False).count()
grouped_likes.head()
```

```python
count_likes_df = grouped_likes[["Venue ID", "Venue"]]
```

```python
merge3_df = merge2_df.merge(count_likes_df, how="inner", on="Venue ID")
merge3_df.head()
```

```python
merge3_df = merge3_df.rename(columns = {"Venue_x":"Venue", "Count":"Tip Count", "Venue_y":"Like Count"})
merge3_df.head()
```

```python
merge3_df.to_csv("merged_data.csv")
```
