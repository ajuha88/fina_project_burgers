```python
import requests, json
import pandas as pd
from config import client_id, client_secret
```

```python
burgers_df = pd.read_csv("burger_joints.csv")
burgers_df = burgers_df.drop("Unnamed: 0", axis=1)
burgers_df.head()
```

```python
CLIENT_ID = client_id
CLIENT_SECRET = client_secret
VERSION = "20180605"
```

```python
all_tips = []

for row in burgers_df.iterrows():
    url =  f'https://api.foursquare.com/v2/venues/{row[1][3]}/tips/'
    
    params = dict(
        client_id=CLIENT_ID,
        client_secret=CLIENT_SECRET,
        v=VERSION,
    )

    response = requests.get(url=url, params=params).json()['response']['tips']['count']
    
    print(f"{row[1][4]}'s Tips: {response}")
    tips = {'count': response}
    
    tips['venue'] = row[1][4]
    tips['venue lat'] = row[1][5]
    tips['venue lon'] = row[1][6]
    tips['venue category'] = row[1][7]

    all_tips.append(tips)
```

```python
tips_df = pd.DataFrame(all_tips)
```
