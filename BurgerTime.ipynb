```python
import requests, json
import pandas as pd
from config import client_id, client_secret
```

```python
hoods = pd.read_csv("hoods.csv")
```

```python
CLIENT_ID = client_id
CLIENT_SECRET = client_secret
VERSION = '20180605'
```

```python
def getBurgers(labels, lats, lons):
    
    burgers = []
    
    for label, lat, lon in zip(labels, lats, lons):
        
        url =  'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&query=burgers&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lon, 
            500, 
            100)
        
        results = requests.get(url).json()['response']['groups'][0]['items']
        
        burgers.append([(
            label,
            lat,
            lon,
            v['venue']['id'],
            v['venue']['name'],
            v['venue']['location']['lat'],
            v['venue']['location']['lng'],
            v['venue']['categories'][0]['name']) for v in results])
    
    return(burgers)
#     burgers_df = pd.DataFrame([burger for burger in burgers])
#     burgers_df.columns = ['Neighborhood', 
#               'Neighborhood Latitude', 
#               'Neighborhood Longitude',
#               'Venue ID',
#               'Venue', 
#               'Venue Latitude', 
#               'Venue Longitude',
#               'Venue Category']

#     return(burgers_df)    
    
```

```python
ny_burgers = getBurgers(labels=hoods['neighborhood'],
                       lats=hoods['latitude'],
                       lons=hoods['longitude'])
```

```python
restaurants = []
for i in ny_burgers:
    for x in i:
        restaurants.append(x)
        
restaurants
```

```python
burgers_df = pd.DataFrame(restaurants)
burgers_df.head()
```

```python
burgers_df.columns = ['Neighborhood', 
              'Neighborhood Latitude', 
              'Neighborhood Longitude',
              'Venue ID',
              'Venue', 
              'Venue Latitude', 
              'Venue Longitude',
              'Venue Category']
burgers_df.head()
```

```python
burgers_df.to_csv("burger_joints.csv")
```
